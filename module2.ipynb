```python
lst = series.tolist()
def standard_deviation(lst):
    average = sum(lst) / len(lst)
    total = 0 
    for number in lst:
        total += (number - average) ** 2
    return (total / (len(lst) - 1)) ** 0.5

print(standard_deviation(lst))
```

Arrange the data in ascending order.

Divide the lowest 25% of the data in the ratio of 1:3 and put the values in Q1.

Divide the remaining 50% of the data in the ratio of 1:1 and put the values in Q2.

Divide the highest 25% of the data in the ratio of 3:1 and the values in Q3.

Put all the values in the formula.

Then calculate the interquartile range based on upper and lower quartile.

```python
def IQR(lst, n):
    lst.sort()
    mid_index = median(lst, 0, n)
    #Q1
    Q1 = lst[median(lst,0,mid_index)]
    #Q3
    Q3 = lst[mid_index + median(lst,mid_index + 1, n)]
    #IQR
    return (Q3-Q1)
    
```
